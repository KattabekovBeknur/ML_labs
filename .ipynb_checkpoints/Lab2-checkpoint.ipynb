```python
import sys
import math as m
import numpy as np
import pandas
import matplotlib.pyplot as plt
import pylab
x = np.linspace(0,15,100)
f = np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2) 
pylab.plot(x,f)
pylab.show()
def func(u):
    return m.sin(u / 5) * m.exp(u / 10) + 5 * m.exp(-u / 2) 
xs1 = [1, 15]
xs2 = [1, 8, 15]
xs3 = [1, 4, 10, 15]

a = np.array([[1,1], [1,15]])
b = np.array([func(xs1[0]), func(xs1[1])])
ans = np.linalg.solve(a, b)
print(str(round(ans[0],2)) +" "+ str(round(ans[1],2)))

Y = ans[0]+ans[1]*x

pylab.plot(x,Y)
pylab.show()
a = np.array([[1,1,1],[1,8,64], [1,15,225]])
b = np.array([func(xs2[0]), func(xs2[1]), func(xs2[2])])
ans = np.linalg.solve(a, b)
print(str(round(ans[0],2)) +" "+ str(round(ans[1],2)) +" " + str(round(ans[2],2)))

Y = ans[0]+ans[1]*x+ans[2]*(x**2)

pylab.plot(x,Y)
pylab.show()
a = np.array([[1,1,1,1],[1,4,16,64],[1,10,100,1000],[1,15,225,3375]])
b = np.array([func(xs3[0]), func(xs3[1]), func(xs3[2]), func(xs3[3])])
ans = np.linalg.solve(a, b)
print(str(round(ans[0],2)) +" "+ str(round(ans[1],2)) +" " + str(round(ans[2],1))+" "+str(round(ans[3],2)))

Y = ans[0]+ans[1]*x+ans[2]*(x**2)+ans[3]*(x**3)

pylab.plot(x,Y)
pylab.show()
```
